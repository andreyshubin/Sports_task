```python
#at first it's necessary to import librares:
#pandas for work with dataframe and scipy for statistics functions

import pandas as pd
from scipy import stats
```

```python
#loading dataset to pandas dataframe 'df'

df = pd.read_csv('dataset_Facebook.csv', sep = ';')
```

```python
#starting aggregation of all dataset with describe method
#it's possible to check mean (the average value), minimum, maximum, values of 25th, 50th, and 75th percentiles
#for each column

df.describe()
```

```python
#we can check parameters with other methods, like

print(df.mean())

#df.median(), df.min(), df.max() and etc
```

```python
#for mode using scipy function, cause built-in pandas (pd.mode()) doesn't work correctly 
#creating new dataframe and add mode and count values 

mode = pd.DataFrame(columns = ['Column name', 'Mode value', 'Count value'])
for column in df:
    mode = mode.append({'Column name':column, 'Mode value':stats.mode(df[column])[0], 'Count value':stats.mode(df[column])[1]}, ignore_index = True)

mode
```

```python
#checking list of 'Type' variables
#then find the same parameters for each object from that list

df.groupby(['Type']).groups.keys()
```

```python
#all necessary stuff

df.groupby('Type').agg('min')
df.groupby('Type').agg('max')
df.groupby('Type').agg('mean')
df.groupby('Type').agg('median')

#in mode example pay attencion - each attribute has 2 numbers: 1 - mode value, 2 - counts value

df.groupby('Type').agg(stats.mode)
```

```python
#for searching most popular object in dataset I was using following parameters:
#total interaction as main, cause amount of comments + shares is very important, likes - too, but with less weight
#and also reach and impressions for cheching amount of views and demonstrations of posts

df.sort_values(['Total Interactions', 'Lifetime Post Total Reach', 'Lifetime Post Total Impressions'], ascending = [False, False, False]).head(1)
```
